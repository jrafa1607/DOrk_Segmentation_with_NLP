```python
#Library Import
import pandas as pd
import numpy as np
import nltk
```

```python
#Path Definition
path = "/content/Ghdb.csv"
```

```python
#ImportDataframe
dataframe = pd.read_csv(path, delimiter=";", encoding= 'unicode_escape', engine='python');
```

```python
#Null Values Total
dataframe.isnull().sum()
```

```python
#Change all null values for "missing"
dataframe.fillna(' ', inplace=True)
```

```python
#Stopwords Definition
stopwords = [',', ':', ';', '"', '-', '=', "’", '!', '?', '”', "'", "''",'(', ')', '``', '@', '~', '/', '|', ' *', '*', '* ', '[,', '],', ',[', ',]', ',[,', ',],', '[', ']', '__', '_', '...', '+', '/\\', '\\', '****', '.', '$', '#', '%', '¨', '¬', '&', '`', './', '/.', './', '/.', '||', '/*', '©'];
```

```python
#Count the total of Atributtes in the Dataframe
linhas = np.size(dataframe,0)
colunas = np.size(dataframe,1)
X = dataframe
```

```python
#Variable Key
maior = 0
```

```python
for n in range (linhas):
    token = X.iloc[n]['Dork']
    token = nltk.word_tokenize(token)

    token = [c for c in token if not c in stopwords]
    token = ''.join(token)

    token = [c for c in token if not c in stopwords]
    token = ''.join(token)

    Tamanho = int(len(token))

    if Tamanho > maior:

        maior = Tamanho
        print(X.iloc[n]['Dork'])
        print(maior)

print(maior)
```

```python
#Full NLP Apply
c = 0

for i in range(maior):

    for n in range (linhas):

        if X.iloc[n,i] == 'Missing':

            Dork = X.iloc[n]['Dork']
            Dork = nltk.word_tokenize(Dork)

            Dork = [c for c in Dork if not c in stopwords]
            Dork = ''.join(Dork)

            Dork = [c for c in Dork if not c in stopwords]

            Dork = ''.join(Dork)
            DorkL = ""

            if len(Dork) > c:

                DorkL = str(ord(str(Dork[c])))
                X.iloc[n,i] = int(DorkL)/100

            else:

                X.iloc[n,i] = 0

    c = c + 1
```
